```python
import numpy as np
import matplotlib.pyplot as plt

#Definieer de straal en het middelpunt van de bol
r = 3  # straal (sqrt(9))
center = (1, -2, 3)  # middelpunt (x0, y0, z0)

#Maak een raster voor de plot
u = np.linspace(0, 2 * np.pi, 100)
v = np.linspace(0, np.pi, 100)

#Bepaal de coördinaten van de bol
X = center[0] + r * np.outer(np.cos(u), np.sin(v))
Y = center[1] + r * np.outer(np.sin(u), np.sin(v))
Z = center[2] + r * np.outer(np.ones(np.size(u)), np.cos(v))

#Plot de bol
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, color='cyan', alpha=0.6)

#Labels en titel
ax.set_xlabel('X-as')
ax.set_ylabel('Y-as')
ax.set_zlabel('Z-as')
ax.set_title('Bol met middelpunt (1, -2, 3) en straal 3')

#Toon de plot
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

#Middelpunt en straal van de bol
center = (1, -2, 3)
radius = np.sqrt(9)

#Maak een grid van punten voor de bol
u = np.linspace(0, 2 * np.pi, 100)  # Hoek in de xy-vlak
v = np.linspace(0, np.pi, 100)       # Hoek in de yz-vlak

#Parametriseer de bol
x = center[0] + radius * np.outer(np.cos(u), np.sin(v))
y = center[1] + radius * np.outer(np.sin(u), np.sin(v))
z = center[2] + radius * np.outer(np.ones(np.size(u)), np.cos(v))

#Coördinaten van de punten A, B, C en D
points = {
    'A': (1, -2, 6),
    
}

#Maak een 3D-plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

#Plot de bol
ax.plot_surface(x, y, z, color='r', alpha=0.6)  # Kleur en transparantie

#Plot de punt A
for point_name, (px, py, pz) in points.items():
    ax.scatter(px, py, pz, label=point_name, s=100)  # s = grootte van de marker

#Labels en titel
ax.set_xlabel('X-as')
ax.set_ylabel('Y-as')
ax.set_zlabel('Z-as')
ax.set_title('3D Bol met middelpunt (1, -2, 3) en straal √9')

#Toon de legenda
ax.legend()

#Toon de plot
plt.show()
```
